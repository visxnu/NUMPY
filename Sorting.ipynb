Sorting

```python
import numpy as np
```

```python
arr=np.array([5,4,8,6,2,9])
print(np.sort(arr))
```

```python
arr=np.array(['bannan','apple','cherry'])
print(np.sort(arr))
```

```python
arr=np.array([[2,1,4],[1,8,3]])
print(arr)
print(np.sort(arr))
arr.sort()
print(arr)
```

```python
arr=np.array([True,False,True])
print(np.sort(arr))
```

                                                            Filtering element

```python
arr=np.array([41,42,43,44,])
x=[True,False,True,False]
newarr=arr[x]
print(newarr)
```

```python
arr=np.random.randint(1,100,40)
filter_array=[]
for element in arr:
    if element>50:
        filter_array.append(True)
    else:
        filter_array.append(False)
new_arr=arr[filter_array]
print(arr)
print(new_arr)
```

```python
arr=np.random.randint(1,100,40)
filter_array=[]
for element in arr:
    if element%2==0:
        filter_array.append(True)
    else:
        filter_array.append(False)
new_arr=arr[filter_array]
print(arr)
print(new_arr)
```

```python
arr=np.random.randint(1,100,40)
filter_array=[]
for element in arr:
    if element%10==0:
        filter_array.append(True)
    else:
        filter_array.append(False)
new_arr=arr[filter_array]
print(arr)
print(new_arr)
```

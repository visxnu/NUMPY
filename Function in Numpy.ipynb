```python
import numpy as np
```

Concatenation

```python
arr1=np.array([[1,2],[3,4]])
arr2=np.array([[5,6],[7,8]])
arr=np.concatenate((arr1,arr2),axis=1)
print(arr)
```

```python
arr1=np.array([[1,2,3],[4,5,6]])
arr2=np.array([[7,8,9],[6,5,3]])
arr=np.concatenate((arr1,arr2),axis=1)
print(arr)
```

Stack function

```python
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])
arr=np.stack((arr1,arr2),axis=0)
print(arr)
```

```python
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])
arr=np.stack((arr1,arr2),axis=1)
print(arr)
```

```python
arr1=np.array([[[1,2,3],[4,5,6]],[[6,7,8],[9,10,11]]])
arr2=np.array([[[7,8,9],[6,5,3]],[[1,2,3],[4,5,6]]])
arr=np.stack((arr1,arr2),axis=3)
print(arr)
```

```python
arr1=np.array([[[1,2,3],[4,5,6]],[[6,7,8],[9,10,11]]])
arr2=np.array([[[7,8,9],[6,5,3]],[[1,2,3],[4,5,6]]])
arr=np.stack((arr1,arr2),axis=0)
print(arr)
```

Different type of stack

```python
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])
arr=np.hstack((arr1,arr2))
print(arr)
```

```python
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])
arr=np.vstack((arr1,arr2))
print(arr)
```

```python
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])
arr=np.dstack((arr1,arr2))
print(arr)
```

```python
import numpy as np
arr=np.zeros(25)
arr=arr.reshape(5,5)
arr[0,:]=1
arr[4,:]=1
arr[:,0]=1
arr[:,4]=1
arr
```

```python
x=np.zeros((8,8))
x[1::2, ::2]=1
x[::2,1::2] =1
print(x)
```

```python
a=np.zeros((4,6))
a[::2,2:4]=1
a[1::2,0:2]=1
a[1::2,4::]=1
a
```

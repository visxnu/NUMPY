```python
import random
win=0
lose=0
tie=0
i=0
while i!=5:
    print("rock-paper-scissor")
    x=input("Enter the choice : ")
    a=random.choice(["rock","paper","scissor"])
    if x==a:
        print("Tied")
        tie=tie+1
    elif x=="rock" and a=="scissor":
        print("You Win")
        win=win+1
    elif x=="paper" and a=="rock":
        print("You Win")
        win=win+1
    elif x=="scissor" and a=="paper":
        print("You Win")
        win=win+1
    else:
        print("You lose")
        lose=lose+1
    i=i+1
print("win=",win)
print("lose=",lose)
print("Tie=",tie)
```

```python
import random
a=100
m=0
while a>=80 and a<=120:
    print("Odd or Even")
    x=input("Enter the choice : ")
    n=random.choice(["odd","even"])
    if x==n:
        print("Win")
        a=a+9
    else:
        print("lose")
        a=a-10
    print("Current point=",a)
print("Game Finished :)")
```

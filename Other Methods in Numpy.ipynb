```python
import numpy as np
```

```python
arr1=np.array([1,2,3])
arr2=np.array([1,2,3])
newarr=np.sum([arr1,arr2])
print(newarr)
```

```python
arr1=np.array([1,2,3])
arr2=np.array([1,2,3])
newarr=np.sum([arr1,arr2],axis=1)
print(newarr)
```

Cummulative sum

It means partially adding the elements in array perform sum with the cumsum() function

```python
arr=np.array([1,2,3])
newarr=np.cumsum(arr)
print(newarr)
```

Numpy products

```python
arr=np.array([[1,2,3],[4,5,6]])
newarr=np.prod(arr)
print(newarr)
```

```python
arr=np.array([1,2,3,6])
arr1=np.array([1,2,3,4])
newarr=np.prod([arr,arr1])
print(newarr)
```

```python
arr=np.array([1,2,3,6])
arr1=np.array([1,2,3,4])
newarr=np.prod([arr,arr1],axis=0)
print(newarr)
```

Cummulative product

```python
arr=np.array([5,6,7,8])
newarr=np.cumprod(arr)
print(newarr)
```

Numpy differences

```python
arr=np.array([10,15,25,5])
newarr=np.diff(arr)
print(newarr)
```

Numpy logs

```python
arr=np.arange(1,10)
print(np.log2(arr))
```

```python
arr=np.arange(1,10)
print(np.log(arr))
```

LCM in numpy

```python
num=4
num1=6
x=np.lcm(num,num1)
print(x)
```

```python
num=5
num1=8
x=np.lcm(num,num1)
print(x)
```

```python
arr=np.array([3,6,9])
x=np.lcm.reduce(arr)
print(x)
```

```python
arr=np.arange(1,11)
x=np.lcm.reduce(arr)
print(x)
```

GDC in NUmpy

```python
num1=6
num2=9
x=np.gcd(num1,num2)
print(x)
```

```python
arr=np.array([20,8,32,36,16])
x=np.gcd.reduce(arr)
print(x)
```

```python
arr=np.arange(1,10)
x=np.gcd.reduce(arr)
print(x)
```

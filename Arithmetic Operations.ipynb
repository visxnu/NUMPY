```python
import numpy as np
```

Addition

```python
arr1=np.array([[1,2,3],[4,5,6]])
arr2=np.array([[7,8,9],[1,2,3]])
new_arr=np.add(arr1,arr2)
print(new_arr)
```

Subtraction

```python
arr1=np.array([[10,17,18],[28,25,6]])
arr2=np.array([[7,8,3],[8,12,2]])
new_arr=np.subtract(arr1,arr2)
print(new_arr)
```

Multipication

```python
arr1=np.array([[1,2,3],[4,5,6]])
arr2=np.array([[7,8,9],[1,2,3]])
new_arr=np.multiply(arr1,arr2)
print(new_arr)
```

Division

```python
arr1=np.array([[1,2,3],[4,5,6]])
arr2=np.array([[7,8,9],[1,2,3]])
new_arr=np.divide(arr1,arr2)
print(new_arr)
```

Transpose of a matric

```python
a=np.array([[[1,2],[3,4]],[[1,3],[2,4]]])
print(a)
print(np.transpose(a))
```

```python
arr=np.array([[2,3,4],[1,7,8]])
print(arr)
print(np.transpose(arr))
```

```python
m=np.array([[1,2,3],[4,5,6]])
n=np.array([[1,7,9],[2,3,4]])
print("addition\n",np.add(m,n))
print("subtraction\n",np.subtract(m,n))
print("multiple\n",np.multiply(m,n))
print("division\n",np.divide(m,n))
print("Transpose of 1st\n",np.transpose(m))
print("Transpose of 2nd\n",np.transpose(n))
```

```python
m=np.array([[1,2],[3,4]])
n=np.array([[4,5],[7,3]])
print(m)
print(n)
print(np.matmul(m,n))
```

```python
a=np.array([[[1,2],[3,4]],[[1,3],[2,4]]])
b=np.array([[[2,3],[2,1]],[[4,5],[8,5]]])
print(np.matmul(a,b))
```
